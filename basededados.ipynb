# PARSER

```python
# Função para atualizar uma célula do daa frame, observando o ano e o código IBGE
def update_cell(df, cod, ano, col, value):
  cod_mask = df["Cod-MUN"] == cod
  ano_mask = df["Ano"] == ano
  idx = df.index[ano_mask & cod_mask]
  df.at[idx, col] = value
```

```python
painel_df = pd.read_csv(data_path + "painel.csv")
painel_df
```

### pop

```python
# pop 2009-2022
br_ibge_populacao_df = pd.read_csv(data_path + "br_ibge_populacao.csv")

# Preenchendo coluna pop
painel_df["pop"] = None

ano_piso = br_ibge_populacao_df["ano"] > 2008

for idx in br_ibge_populacao_df[ano_piso].index:
  ano = br_ibge_populacao_df.at[idx, "ano"]
  cod = br_ibge_populacao_df.at[idx, "id_municipio"]
  pop = br_ibge_populacao_df.at[idx, "populacao"]
  update_cell(painel_df, cod, ano, "pop", pop)
```

```python
painel_df.to_csv(data_path + "painel.csv", index=False)
```

### IDHM

```python
# Coluna IDHM
mundo_onu_adh_df = pd.read_csv(data_path + "mundo_onu_adh.csv")

painel_df["IDHM"] = None

ano_limite = mundo_onu_adh_df["ano"] == 2010

for i in mundo_onu_adh_df[ano_limite].index:
  ano = 2010
  cod = mundo_onu_adh_df.at[i, "id_municipio"]
  value = mundo_onu_adh_df.at[i, "idhm"]
  update_cell(painel_df, cod, ano, "IDHM", value)
```

### colunas: idbeAI, idbeAF, idbeEM

```python
br_inep_ideb_df = pd.read_csv(data_path + "br_inep_ideb.csv")
```

```python
painel_df["idbeAI"] = None
painel_df["idbeAF"] = None
painel_df["idbeEM"] = None

ano_limite = br_inep_ideb_df["ano"] > 2008
ensino_medio = br_inep_ideb_df["ensino"] == "medio"
ensino_fundamental = br_inep_ideb_df["ensino"] == "fundamental"
anos_iniciais = br_inep_ideb_df["anos_escolares"] == "iniciais (1-5)"
anos_finais = br_inep_ideb_df["anos_escolares"] == "finais (6-9)"
publica = br_inep_ideb_df["rede"] == "publica"

#idbeEM
for i in br_inep_ideb_df[ensino_medio & ano_limite & publica].index:
  ano = br_inep_ideb_df.at[i, "ano"]
  cod = br_inep_ideb_df.at[i, "id_municipio"]
  value = br_inep_ideb_df.at[i, "ideb"]
  update_cell(painel_df, cod, ano, "idbeEM", value)

#idbeAF
for i in br_inep_ideb_df[anos_iniciais & ano_limite & ensino_fundamental & publica].index:
  ano = br_inep_ideb_df.at[i, "ano"]
  cod = br_inep_ideb_df.at[i, "id_municipio"]
  value = br_inep_ideb_df.at[i, "ideb"]
  update_cell(painel_df, cod, ano, "idbeAI", value)

#idbeAI
for i in br_inep_ideb_df[anos_finais & ano_limite & ensino_fundamental & publica].index:
  ano = br_inep_ideb_df.at[i, "ano"]
  cod = br_inep_ideb_df.at[i, "id_municipio"]
  value = br_inep_ideb_df.at[i, "ideb"]
  update_cell(painel_df, cod, ano, "idbeAF", value)
```

### colunas: dist_isef_[2014-2020], dist_ai_[2014-2020], dist_af_[2014-2020], dist_em_[2014-2020]

```python
br_inep_indicadores_educacionais_df = pd.read_csv(data_path + "br_inep_indicadores_educacionais.csv")
painel_df = pd.read_csv(data_path + "painel.csv")

painel_df["Dist_ISEF"] = None
painel_df["Dist_AI"] = None
painel_df["Dist_AF"] = None
painel_df["Dist_EM"] = None

ano_limite = br_inep_indicadores_educacionais_df["ano"] > 2008
loc_total = br_inep_indicadores_educacionais_df["localizacao"] == "total"
rede_total = br_inep_indicadores_educacionais_df["rede"] == "total"

for i in br_inep_indicadores_educacionais_df[loc_total & ano_limite & rede_total].index:
  ano = br_inep_indicadores_educacionais_df.at[i, "ano"]
  cod = br_inep_indicadores_educacionais_df.at[i, "id_municipio"]
  value_isef = br_inep_indicadores_educacionais_df.at[i, "tdi_ef"]
  value_ai = br_inep_indicadores_educacionais_df.at[i, "tdi_ef_anos_iniciais"]
  value_af = br_inep_indicadores_educacionais_df.at[i, "tdi_ef_anos_finais"]
  value_em = br_inep_indicadores_educacionais_df.at[i, "tdi_em"]
  update_cell(painel_df, cod, ano, "Dist_ISEF", value_isef)
  update_cell(painel_df, cod, ano, "Dist_AI", value_ai)
  update_cell(painel_df, cod, ano, "Dist_AF", value_af)
  update_cell(painel_df, cod, ano, "Dist_EM", value_em)

  painel_df.to_csv(data_path + "painel.csv", index=False)
```
